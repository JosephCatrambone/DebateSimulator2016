```python
fout = open("presidential_alignment.csv", 'wt')
cout = csv.writer(fout)
cout.writerow([x.lower() for x in header + ['words']])
for p in presidents:
    # Build the row.
    row = list()
    for h in header:
        row.append(p[h])
    pres_lastname = p['PRESIDENT'].split()[-2]
    # Get the speeches
    words = set()
    for f in filenames:
        if f.startswith("Corpus of Presential Speeches/{}".format(pres_lastname.lower())):
            pres_speech = zf.open(f).read()
            for w in pres_speech.lower().split():
                clean_word = "".join([chr(c) for c in w if w.isalpha()])
                words.add(clean_word)
    row.append("|".join(words))
    cout.writerow(row)
fout.close()
```
